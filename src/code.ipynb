```python
uk_ranking = ranking_df[(ranking_df['Year'] == 2025) & 
                        (ranking_df['Country'] == 'United Kingdom')]
```

```python
uk_ranking.head(10)
```

```python
uk_ranking.loc[:, 'Name'] = uk_ranking['Name'].str.strip()
uk_ranking.loc[:, 'Country'] = uk_ranking['Country'].str.strip()
uk_cs_courses.loc[:, 'University'] = uk_cs_courses['University'].str.strip()
uk_cs_courses.loc[:, 'Country'] = uk_cs_courses['Country'].str.strip()
```

```python
merged = pd.merge(
    uk_cs_courses,
    uk_ranking,
    left_on=['University'],
    right_on=['Name'],
) 
```

```python
merged.head(10)
```

```python
merged_sorted = merged.sort_values('Tuition_rate')
merged_sorted[['Rank','Name','Year','Program','Tuition_rate','Avg_Uni_Tuition']].sort_values(by='Rank', ascending=True)
```

```python
uni_dict = {'UCL':'University College London'}
uni_entry['Name']=uni_entry['Name'].map(ucl_dict)
```

```python
uk_ranking = uk_ranking.copy()
```

```python
def cast_uni_name(name):
    if name in uni_dict:
        return uni_dict[name]
    else:
        return name
```

```python
uk_ranking['Name']=uk_ranking['Name'].apply(cast_uni_name)
```

```python
uk_ranking
```

```python
merged = pd.merge(
    uk_cs_courses,
    uk_ranking,
    left_on=['University'],
    right_on=['Name'],
)
```

```python
merged_sorted = merged.sort_values('Tuition_rate')
merged_sorted[['Rank','Name','Year','Program','Tuition_rate','Avg_Uni_Tuition']].sort_values(by='Rank', ascending=True)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=merged_sorted, x='Rank', y='Avg_Uni_Tuition', hue='Name')
plt.title('Average University Tuition vs. University Rank')
plt.xlabel('Rank')
plt.ylabel('Average University Tuition')
plt.gca().invert_xaxis()  # Lower rank means better university, so invert for visual logic
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.show()
```
